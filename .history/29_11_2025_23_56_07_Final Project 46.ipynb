Start by loading `players.csv` into an object called `players`.

Because the K-nearest neighbors model uses `Age` to predict `subscribe`, only these two variables should be present within the dataframe. In addition, since k-nn regression can only be used to predict categorical data, `subscribe` must be converted from a logical type to a factor type.

```r
players <- players |>
    select(subscribe, Age) |>
    mutate(subscribe = as_factor(subscribe))
players
```
